# Data Visualization

## Area Plot

#### Bagaimana Distribusi peserta terhadap Kategori E-Learning setiap bulannya?

```python
feedback_elearning.head()
```

```python
#melihat daftar category e-learning yang ada
jenis_kategori = feedback_elearning['Category'].unique()
jenis_kategori
```

```python
#menghitung jumlah peserta per katergori, urut dengan bulan terkait
total_peserta_kategori = feedback_elearning[['Bulan','Category','Jumlah Partisipan']]
total_peserta_kategori
```

```python
#grouping total peserta kedalam masing-masing kategori
total_peserta_kategori = total_peserta_kategori.groupby(['Bulan', 'Category'])['Jumlah Partisipan'].sum().reset_index()
total_peserta_kategori
```

```python
#pivoting data peserta per kategori dengan bulan sebagai acuannya
total_peserta_kategori = total_peserta_kategori.pivot(index='Bulan', columns='Category', values='Jumlah Partisipan')
total_peserta_kategori
```

```python
#visualisasi data diatas kedalam are plot

color_list = ['gold','yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink','coral', 'darkslateblue', 'mediumseagreen']
total_peserta_kategori.plot(kind='area',
                            figsize=(20,10),
                           alpha=0.5,
                            stacked=False,
                           color=color_list,
                           )
plt.title('Trend Pengisian Feedback selama 2020') #posisi title
plt.ylabel('Jumlah Partisipan')
plt.xlabel('Bulan')

plt.show()
```

### Analisis
Terlihat pada grafik diatas bahwa distribusi jumlah peserta terbanyak terjadi pada kategori "top & new courses" dimulai dari bulan ke-5 dan seterusnya. ini cukup make sense dimana pada kategori ini elearning yang diberikan kepada peserta termasuk kedalam 'Mandatory E-learning' sehingga grafiknya meningkat tajam.

## PIE CHART
bagaimana persentase jumlah partisipan terhadap top 5 kategori?

```python
#mengambil kolom kategori dan partisipan sebagai data awal
top5_kategori = feedback_elearning[['Category','Jumlah Partisipan']]
```

```python
# grouping berdasar kategori
top5_kategori = top5_kategori.groupby(['Category'], axis=0).sum()
```

```python
#sorting
top5_kategori.sort_values(by='Jumlah Partisipan',ascending=False, inplace=True)
```

```python
top5_kategori = top5_kategori.head(5)
top5_kategori
```

```python
color_list = ['gold','yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink','coral', 'darkslateblue', 'mediumseagreen']

# ratio for each continent with which to offset each wedge.
explode_list = [0, 0, 0,0.1,0.1] 

# setting plot for sorted category peserta
top5_kategori['Jumlah Partisipan'].plot(kind='pie',
                            autopct='%1.1f%%',  #nampilin persentase
                            figsize=(15,6),     #ukuran pie chart
                            startangle=90,      #posisi 90 derajat
                            shadow=True,        #biar ada bayangan
                            counterclock=False, #searah jarum jam
                            labels=None,        #ngilangin label
                            pctdistance=1.12,   #jarak persentase dari pie chartnya
                            colors=color_list,
                            explode=explode_list
                           )
plt.title('Top 5 Category', y = 1.12) #posisi title
plt.axis('equal') #buat posisi pie chart ditengah
plt.legend(labels=top5_kategori.index, loc='upper left') #posisi legend pakai LOC
plt.show()
```

#### Analisis
2 porsi terbesar dalam grafik didominasi oleh "Top & New COurse" dan " DIgital & ICT (EL-DIC) dimana konten yang terkandung didalamnya tidak hanya bersifat mandatory, namun juga perubahan zaman industri 4.0, kebanyakan peserta tertarik untuk mempelajari tentang ilmu yang bersifat baru dan dunia digital

## BOX PLOT
bagaiman distribusi peserta di top 3 job family selama tahun 2020?

```python
#melakukan pemilihan kolom tertentu, grouping, dan pivoting dengan bulan sebagai indexnya
total_peserta_kategori = feedback_elearning[['Bulan','Category','Jumlah Partisipan']]
total_peserta_kategori = total_peserta_kategori.groupby(['Bulan', 'Category'])['Jumlah Partisipan'].sum().reset_index()
total_peserta_kategori = total_peserta_kategori.pivot(index='Bulan', columns='Category', values='Jumlah Partisipan')
total_peserta_kategori
```

```python
#memilih top 3 Jumlah partisipan dari kategori Job Family tertentu
top3_jobfam = total_peserta_kategori[['Business Enabler / Enterprise Management (EL-EEM)', 'Commerce & Relations (EL-COR)',
       'Digital & ICT (EL-DIC)'
]]
top3_jobfam
```

```python
#create box plot
top3_jobfam.plot(kind='box',
                 figsize=(10,6),
                 vert=False
                )
plt.title('Box Plot top 3 job family selama setahun')
plt.xlabel('Jumlah Partisipan')
plt.ylabel('Job Family')

plt.show(20,20)
```

#### Analisis
bahwa Job Family Digital & ICT merupakan job family dengan jumlah partisipan terbanyak, sangat masuk akal karena konten yang terkandung di elarningnya bersifat mandatory dan mengikuti strategi perusahaan untuk berubah menjadi digital telelkomunikasi kedepannya

## WORD CLOUD

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# define stopwords
stopwords = set(STOPWORDS)
#pemilihan kolom yang akan diperiksa
nama_course = feedback_elearning['Course']
nama_course = feedback_elearning['Course'].to_string()
```

```python
wc_course = WordCloud(
    background_color='white',
    stopwords=stopwords
)
#generate wordcloud
wc_course.generate(nama_course)
```

```python
fig = plt.figure()
fig.set_figwidth(12)
fig.set_figheight(12)

plt.imshow(wc_course,interpolation='bilinear')
plt.axis('off')
plt.show()
```

#### Analisis
Terlihat bahwa Kolom "course" kata kata seperti 'Product Knowledge', 'CX Transformation', 'TCP IP', 'Transformation Program' dan 'Communication Skill' adalah kata yang cukup sering muncul
