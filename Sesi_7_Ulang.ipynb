## Visualizing Data Using Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

```python
haiti = df_can.loc['Haiti', years]
```

```python
haiti.head()
```

### LINE PLOT

```python
haiti.plot()
```

```python
haiti.plot(kind='line')
plt.title('Imigration from Haiti')
plt.ylabel('Number of imigration')
plt.xlabel('Years')

plt.show()
```

```python
haiti.index = haiti.index.map(int)
```

```python
haiti.plot(kind='line')
plt.title('Imigration from Haiti')
plt.ylabel('Number of imigration')
plt.xlabel('Years')

plt.text(2000,6000,'2010 Earthquake')

plt.show()
```

```python
China_India = df_can.loc[['China', 'India'], years]
```

```python
China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()
```

```python
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Imigration from China and India')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')
#plt.text(2000,4000, '2005 Earthquake')
plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False,axis=0, inplace=True)
```

```python
df_top5 = df_can.head()
```

```python
df_top5
```

```python
df_top5 = df_top5[years].transpose()
```

```python
df_top5.plot(kind='line', figsize=(14,9))
plt.title('Imigration Trend of top 5 Countries')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')
#plt.text(2000,4000, '2005 Earthquake')
plt.show()
```

### AREA PLOT

```python
df_top5.head()
```

```python
df_top5.plot(kind='area', figsize=(20,10), stacked=False, alpha=0.25)
plt.title('Imigration Trend of top 5 Countries')
plt.ylabel('Number of imigrants')
plt.xlabel('Years')
#plt.text(2000,4000, '2005 Earthquake')
plt.show()
```
