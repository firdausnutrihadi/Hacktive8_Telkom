### PIE CHART

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                            autopct='%1.1f%%', #nampilin persentase
                            figsize=(10,6),
                           startangle=90,      #posisi Afrika 90 derajat
                            shadow=True,
                            counterclock=False #searah jarum jam
                           )
plt.title('Immigration to Canada by Continent')
plt.axis('equal') #buat posisi pie chart ditengah

plt.show()
```

```python
color_list = ['gold','yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1,0,0,0,0.1,0.1] #keluarin pie nya dari lingkaran
df_continents['Total'].plot(kind='pie',
                            autopct='%1.1f%%',  #nampilin persentase
                            figsize=(15,6),     #ukuran pie chart
                            startangle=90,      #posisi Afrika 90 derajat
                            shadow=True,        #biar ada bayangan
                            counterclock=False, #searah jarum jam
                            labels=None,        #ngilangin label
                            pctdistance=1.12,   #jarak persentase dari pie chartnya
                            colors=color_list,
                            explode=explode_list
                           )
plt.title('Immigration to Canada by Continent', y = 1.12) #posisi title
plt.axis('equal') #buat posisi pie chart ditengah
plt.legend(labels=df_continents.index, loc='upper left') #posisi legend pakai LOC
plt.show()
```

```python
color_list = ['gold','yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink']
explode_list = [0.1,0,0,0,0.1,0.1] #keluarin pie nya dari lingkaran
df_continents['2013'].plot(kind='pie',
                            autopct='%1.1f%%',  #nampilin persentase
                            figsize=(15,6),     #ukuran pie chart
                            startangle=90,      #posisi Afrika 90 derajat
                            shadow=True,        #biar ada bayangan
                            counterclock=False, #searah jarum jam
                            labels=None,        #ngilangin label
                            pctdistance=1.12,   #jarak persentase dari pie chartnya
                            colors=color_list,
                            explode=explode_list
                           )
plt.title('Immigration to Canada by Continent Year 2013', y = 1.12) #posisi title
plt.axis('equal') #buat posisi pie chart ditengah
plt.legend(labels=df_continents.index, loc='upper left') #posisi legend pakai LOC
plt.show()
```

### SUBPLOT

```python
fig = plt.figure()

ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

#subplot 1 : Box plot 
df_CI.plot(kind='box',
          figsize=(20,6),
          vert=False,
          color='skyblue',
          ax=ax0)
ax0.set_title('Box plot of China and India')
ax0.set_ylabel('number of immigrants')
ax0.set_xlabel('countries')

#subplot 2 : Line plot
df_CI.plot(kind='line',
       figsize=(20,6),
       ax=ax1
        )
ax1.set_title('Line Plot of China and India')
ax1.set_ylabel('number of immigrants')
ax1.set_xlabel('countries')

plt.show()
```

### SCATTER PLOT

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
```

```python
df_tot.index = map(int, df_tot.index)
```

```python
df_tot.reset_index(inplace=True)
```

```python
df_tot.columns=['year', 'total']
```

```python
df_tot.head()
```

```python
df_tot.plot(kind='scatter',
           x='year',
            y='total',
            figsize=(10,6),
            color='darkblue'
           )
plt.title('Total imigration to Canada') #posisi title
plt.xlabel('Years')
plt.ylabel('Total Number of Immigrants')

plt.show()
```
